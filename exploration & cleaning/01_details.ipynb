## 1. Import e caricamento dati
Importiamo le librerie necessarie e carichiamo il file csv. Facciamo una esplorazione generica per capire la struttura e le caratteristiche del dataset.

```python
import pandas as pd
import numpy as np
from dataset_analyzer import analyze

df_details = pd.read_csv('../datasets/details.csv')
print(f'Shape: {df_details.shape}')
print()
df_details.info()
df_details.head()
```

Il dataset contiene **28.955 righe** e **29 colonne**. Alcune colonne presentano valori nulli. Alcuni dtype richiedono conversione: le date sono in formato stringa, e alcune colonne intere (`scored_by`, `rank`, `episodes`, `year`) sono memorizzate come `float64`. Quiesti veranno analizzati e gestiti singolarmente nella sezione 2. 

## 1.1 Rimozione duplicati esatti

Prima dell'analisi per colonna, rimuoviamo le righe con valori identici in **tutte** le colonne, mantenendo solo la prima occorrenza.

```python
n_originale = len(df_details)

mask_dup = df_details.duplicated(keep=False)       
n_righe_coinvolte = mask_dup.sum()                  
n_gruppi = df_details[mask_dup].duplicated(keep='first').sum()  
n_tenute = n_righe_coinvolte - n_gruppi            

print(f'Righe totali coinvolte in duplicazioni : {n_righe_coinvolte:,}')
print(f'  → prime occorrenze mantenute         : {n_tenute:,}')
print(f'  → occorrenze extra rimosse           : {n_gruppi:,}')
print()

df_details.drop_duplicates(keep='first', inplace=True)
print(f'Righe prima della rimozione : {n_originale:,}')
print(f'Righe dopo la rimozione     : {len(df_details):,}')
```

**Verifica consistenza `start_date` ↔ `end_date`**

Una volta convertite entrambe le date, verifichiamo che non ci siano righe in cui `end_date < start_date`: un anime non può terminare prima di iniziare. Consideriamo solo le righe in cui entrambe le date sono presenti, poiché confrontare con un valore nullo non ha significato.

```python
# end_date < start_date
both_valid = df_details['start_date'].notna() & df_details['end_date'].notna()
inverted   = both_valid & (df_details['end_date'] < df_details['start_date'])
print(f'Righe con end_date < start_date: {inverted.sum()}')
print()
print(df_details[inverted][['mal_id', 'title', 'status', 'start_date', 'end_date']].head(10).to_string(index=False))
```

**Osservazioni:**
- **101 righe** presentano `end_date < start_date`. Guardando i dati, il pattern è sempre lo stesso: `end_date` è il 1° gennaio di un anno (es. `2020-01-01`), mentre `start_date` è una data più tarda nello stesso anno (es. `2020-03-01`).
- La spiegazione è che a volte quando si conosce solo l'**anno** di fine trasmissione, non il giorno esatto si registra automaticamente il 1° gennaio come data di fine. Il risultato è che la data di fine appare prima della data di inizio, ma non è un errore reale, è solo una mancanza di precisione nel dato originale. Facciamo una ulteriore verifica per vedere se ci sono casi che rispettano questo pattern. 

```python
# Righe con end_date < start_date e end_date != 1° gennaio
both_valid = df_details['start_date'].notna() & df_details['end_date'].notna()
inverted   = both_valid & (df_details['end_date'] < df_details['start_date'])
not_jan1   = inverted & ~((df_details['end_date'].dt.month == 1) & (df_details['end_date'].dt.day == 1))
print(f'Righe con end_date < start_date e end_date != 1° gennaio: {not_jan1.sum()}')
print()
print(df_details[not_jan1][['mal_id', 'title', 'status', 'start_date', 'end_date']].to_string(index=False))
```

Abbiamo effettuato una ricerca e risulta che per questi anime non si conosce la data esatta di fine trasmissione. Siccome si conosce solo l'anno, li normaliziamo per corrispondere al resto dove la data di fine è impostata a 01 Gennaio dello stesso anno. C'è un ecezzione dove l'anime `Tobot: Daedosiui Yeongungdeul Season 2 Part 2` sembra essere terminato 1 anno prima della data di rilascio. Abbiamo verificato che si tratta di un errore di battitura. Anche questo è terminato nello stesso anno. 

```python
both_valid = df_details['start_date'].notna() & df_details['end_date'].notna()
inverted   = both_valid & (df_details['end_date'] < df_details['start_date'])
not_jan1   = inverted & ~((df_details['end_date'].dt.month == 1) & (df_details['end_date'].dt.day == 1))

df_details.loc[not_jan1, 'end_date'] = df_details.loc[not_jan1, 'start_date'].apply(
    lambda dt: dt.replace(month=1, day=1)
)

print(f'end_date corrette: {not_jan1.sum()}')
print()
print(df_details.loc[not_jan1.index[not_jan1], ['mal_id', 'title', 'start_date', 'end_date']].to_string(index=False))
```

### 2.12 `synopsis`

```python
df_details['synopsis'] = df_details['synopsis'].str.strip()
analyze(df_details['synopsis'])
```

**Osservazioni:**
- **5.127 valori nulli** (~17.7%): strutturali, potrebbe trattarsi di cortometraggi, CM (Commercial Message) e PV (Promotional Video) che spesso non hanno una sinossi registrata.

**Duplicati nella sinossi:**

`analyze()` riporta 288 valori duplicati. Esaminiamo i testi più ricorrenti per verificare se esistono altri pattern boilerplate non catturati dal filtro principale, oppure se si tratta di duplicati legittimi (sinossi condivise tra stagioni o titoli correlati).

```python
# Duplicati in synopsis (su dati non ancora puliti)
dup_mask_syn = df_details['synopsis'].duplicated(keep=False) & df_details['synopsis'].notna()
dup_counts = df_details.loc[dup_mask_syn, 'synopsis'].value_counts()

print(f'Righe con synopsis duplicata: {dup_mask_syn.sum()}')
print(f'Testi univoci tra i duplicati: {len(dup_counts)}')
print()

top15 = dup_counts.head(15).reset_index()
top15.columns = ['synopsis', 'count']
top15['preview'] = top15['synopsis'].str[:120].str.replace('\n', ' ', regex=False).apply(repr)
top15['line'] = top15['count'].astype(str).str.rjust(3) + 'x  ' + top15['preview']

print('Top 15 synopsis duplicate (prime 120 caratteri):')
print('─' * 80)
print('\n'.join('  ' + top15['line']))
```

**Risultati indagine duplicati:**

- **456 righe coinvolte** in duplicazioni (168 testi univoci distinti). Il numero sembra diverso dal numero 288 riportato da `analyze()` perché quella cifra conta solo le copie non-prime, mentre qui si contano tutte le righe con testo non unico.
- Il testo più frequente (23×) è il boilerplate `"No synopsis has been added for this series yet..."` che rimuoviamo successivamente.
- Tutti gli altri duplicati sono descrizioni brevi e generiche di cortometraggi d'autore o descrizioni standard di release. Si tratta di contenuto corretto.
- **Nessun nuovo pattern boilerplate** da rimuovere. I duplicati rimanenti sono legittimi.

**Sinossi contenenti URL e `@`:**

`analyze()` riporta 3 sinossi con URL e 38 con `@`. Verifichiamo il contesto per determinare se si tratta di contenuto legittimo o artefatti da rimuovere.

```python
import re

# Sinossi contenenti URL
url_mask_syn = df_details['synopsis'].str.contains(r'https?://\S+', regex=True, na=False)
print(f'Sinossi con URL: {url_mask_syn.sum()}')
print()

df_url = df_details.loc[url_mask_syn, ['mal_id', 'title', 'type']].copy()
df_url['synopsis'] = df_details.loc[url_mask_syn, 'synopsis'].str.findall(r'https?://\S+')
df_url = df_url.explode('synopsis').reset_index(drop=True)
print(df_url.to_string(index=False))

print()
print('─' * 80)
print()

# Sinossi contenenti '@'
at_mask_syn = df_details['synopsis'].str.contains(r'@', regex=False, na=False)
print(f'Sinossi con @: {at_mask_syn.sum()}')
print()

df_at = df_details.loc[at_mask_syn, ['mal_id', 'title']].copy()
df_at['synopsis'] = df_details.loc[at_mask_syn, 'synopsis'].str.findall(r'\S*@\S+')
df_at = df_at.explode('synopsis').reset_index(drop=True)
print(df_at.to_string(index=False))
```


- Tutte le sinossi contenenti URL sono citazioni di fonte. 
- In tutti i casi contenenti `@`, si tratta di un carattere intenzionale di titoli o nomi ufficiali. 

**Nessuna pulizia necessaria.**

```python
pattern = r'^No synopsis has been added for this series yet'
mask_bp = df_details['synopsis'].str.contains(pattern, regex=True, na=False)
print(f'Righe con sinossi boilerplate: {mask_bp.sum()}')
print(df_details.loc[mask_bp, 'synopsis'].value_counts().to_string())

df_details.loc[mask_bp, 'synopsis'] = np.nan
print(f'\nNull in synopsis dopo pulizia: {df_details["synopsis"].isna().sum():,}')
```

- Effetuiamo un controllo sulla coerenza tra `start_date` e `year`. 

```python
df_details['year'] = df_details['year'].astype('Int64')
year_from_date = df_details["start_date"].dt.year
mismatch = df_details["year"].notna() & df_details["start_date"].notna() & (df_details["year"].astype("Int64") != year_from_date.astype("Int64"))
print(f"year != start_date.year: {mismatch.sum()}")
print()
print(df_details[mismatch][["mal_id", "title", "year", "season", "start_date"]].head(10).to_string(index=False))
```

La stagione invernale corrisponde al periodo dicembre-marzo, quindi un anime classificato come "winter 2021" puo avere `start_date` a dicembre 2020. Il `year` riporta 2021 (anno della stagione MAL) mentre `start_date.year` riporta 2020 (anno della messa in onda effettiva). La discrepanza e attesa e coerente.
