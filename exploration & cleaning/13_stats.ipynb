# 13 – Stats

Esplorazione e data cleaning del dataset `stats.csv`.

**Colonne:**
| Colonna | Descrizione |
|---|---|
| `mal_id` | ID dell'anime su MAL (PK, FK → `details.csv`) |
| `watching` | Numero di utenti che stanno guardando l'anime |
| `completed` | Numero di utenti che hanno completato l'anime |
| `on_hold` | Numero di utenti che hanno messo l'anime in pausa |
| `dropped` | Numero di utenti che hanno abbandonato l'anime |
| `plan_to_watch` | Numero di utenti che intendono guardare l'anime |
| `total` | Totale utenti che hanno l'anime in lista |
| `score_N_votes` | Numero di voti ricevuti per il punteggio N (1–10) |
| `score_N_percentage` | Percentuale di voti per il punteggio N rispetto al totale dei voti |

## 1. Import e caricamento dati

```python
import pandas as pd
import numpy as np
from dataset_analyzer import analyze
```

```python
df_stats = pd.read_csv('../datasets/stats.csv')
print(f'Shape: {df_stats.shape}')
df_stats.info()
df_stats.head()
```

```python
n_originale = len(df_stats)
```

## 1.1 Verifica duplicati esatti

Verifichiamo la presenza di righe completamente identiche.

```python
n_dup = df_stats.duplicated().sum()
print(f'Duplicati esatti: {n_dup}')

if n_dup > 0:
    df_stats.drop_duplicates(keep='first', inplace=True)
    print(f'Righe dopo la rimozione: {len(df_stats):,}')
else:
    print('→ Nessun duplicato esatto, nessuna operazione richiesta.')
```

## 2. Analisi colonna per colonna

### 2.1 `mal_id`

```python
analyze(df_stats['mal_id'])
```

**Nessuna pulizia necessaria.**  
Nessun null, dtype già `int64`, nessun duplicato. È la chiave primaria del dataset: ogni riga corrisponde a un anime distinto. Chiave esterna verso `details.csv`.

```python
print(f'Null in mal_id       : {df_stats["mal_id"].isna().sum()}')
print(f'Duplicati in mal_id  : {df_stats["mal_id"].duplicated().sum()}')
print(f'Anime distinti       : {df_stats["mal_id"].nunique():,}')
```

### 2.2 Colonne di conteggio: `watching`, `completed`, `on_hold`, `dropped`, `plan_to_watch`, `total`

```python
count_cols = ['watching', 'completed', 'on_hold', 'dropped', 'plan_to_watch', 'total']
for col in count_cols:
    analyze(df_stats[col])
```

**Nessuna pulizia necessaria.**  
Nessun null in nessuna delle colonne di conteggio, dtype già `int64`. I valori sono coerenti: ogni conteggio è ≥ 0 e `total` corrisponde alla somma delle cinque categorie.

```python
print('Null nelle colonne di conteggio:')
print(df_stats[count_cols].isnull().sum())
print()
print('Statistiche descrittive:')
print(df_stats[count_cols].describe())
print()
# Verifica coerenza: total == somma delle 5 categorie
somma_categorie = df_stats[['watching', 'completed', 'on_hold', 'dropped', 'plan_to_watch']].sum(axis=1)
discrepanze = (somma_categorie != df_stats['total']).sum()
print(f'Righe dove total ≠ somma categorie: {discrepanze:,}')
```

### 2.3 Colonne `score_N_votes` e `score_N_percentage` (N = 1–10)

Tutte e 20 le colonne presentano lo stesso numero di null (430): si tratta delle stesse righe, ovvero anime che non hanno ancora ricevuto voti su MAL. Il null è strutturale e va mantenuto.

```python
score_votes_cols = [f'score_{n}_votes' for n in range(1, 11)]
score_pct_cols   = [f'score_{n}_percentage' for n in range(1, 11)]
score_cols       = score_votes_cols + score_pct_cols

for col in score_cols:
    analyze(df_stats[col])
```

**Pulizia necessaria:**  
- Le colonne `score_N_votes` hanno dtype `float64` a causa dei 430 null; poiché rappresentano conteggi interi, vengono convertite in `Int64` (intero nullable di pandas).
- Le colonne `score_N_percentage` rimangono `float64` (valori percentuali).
- I 430 null sono strutturali: le stesse righe mancano di dati per tutti e 20 i campi (anime senza nessun voto registrato su MAL). Non vengono rimossi.

```python
# Verifica che i null siano nelle stesse righe per tutti i campi score
mask_no_score = df_stats['score_1_votes'].isna()
tutti_stessi = all(df_stats[col].isna().equals(mask_no_score) for col in score_cols)
print(f'Righe senza dati di voto: {mask_no_score.sum():,}')
print(f'Null nelle stesse righe per tutti i campi score: {tutti_stessi}')
print()
print('Caratteristiche anime senza score (colonne di conteggio):')
print(df_stats[mask_no_score][['watching', 'completed', 'on_hold', 'dropped', 'plan_to_watch', 'total']].describe())
```

```python
# Conversione score_N_votes da float64 a Int64 (nullable integer)
for col in score_votes_cols:
    df_stats[col] = df_stats[col].astype('Int64')

print('Dtype score_N_votes dopo la conversione:')
print(df_stats[score_votes_cols].dtypes)
```

## 3. Data Cleaning

```python
print('=== Riepilogo Dataset Pulito ===')
print(f'Righe originali      : {n_originale:>12,}')
print(f'Righe dopo cleaning  : {len(df_stats):>12,}')
print(f'Righe rimosse totali : {n_originale - len(df_stats):>12,}')
print()
print('Dtypes finali:')
print(df_stats.dtypes)
print()
print('Valori mancanti residui (strutturali):')
print(df_stats.isnull().sum())
```
